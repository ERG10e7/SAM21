```python
from matplotlib import pyplot as plt
import numpy as np

data = np.genfromtxt("CG20.dat", delimiter=",", names = True)

fig = plt.figure(figsize=(10,10))

fig.subplots_adjust(hspace=0.3, wspace=0.5)

ax1 = fig.add_subplot(221)
ax1.scatter(data['GLON'],data['GLAT'],s=50*data['r50'],c=data['AgeNN'],alpha=0.3)  
ax1.set_xlabel("Galactic Longtude",size=15)
ax1.set_ylabel("Galactic Latitude",size=15)

image = ax1.scatter(data['X'],data['Y'],s=200*data['r50'],c=data['AgeNN'],cmap="viridis",alpha=0.5)
bar = fig.colorbar(image,orientation="vertical",pad=0.01)
bar.set_label('$log_{10} (Age)$',size=15)

ax2 = fig.add_subplot(222)
ax2.scatter(data['pmRA'],data['pmDE'],s=50*data['r50'],c=data['AgeNN'],alpha=0.3)  
ax2.set_xlabel("Right Ascention",size=20)
ax2.set_ylabel("Declination",size=20)


bar = fig.colorbar(image,orientation="vertical",pad=0.01)
bar.set_label('$log_{10} (Age)$',size=15)

ax3 = fig.add_subplot(223)
ax3.scatter(data['Rgc'],data['Z'],s=50*data['r50'],c=data['AgeNN'],alpha=0.3)  
ax3.set_xlabel("Radius from G.Center",size=20)
ax3.set_ylabel("Z coordinate",size=20)

bar = fig.colorbar(image,orientation="vertical",pad=0.01)
bar.set_label('$log_{10} (Age)$',size=15)


ax3_sub = fig.add_axes([0.278,0.155,0.1,0.1]) #right,up,hz,vert
ax3_sub.hist(data['nbstars07'],bins=np.logspace(0,3,20),log='True')
ax3_sub.set_xscale('log')
ax3_sub.set_ylabel('$log_{10} (Stars)$', size=10)


ax4 = fig.add_subplot(224)
ax4.scatter(data['AgeNN'],data['DistPc'],s=50*data['r50'],c=data['AgeNN'],alpha=0.3)  
ax4.set_xlabel("Age",size=20)
ax4.set_ylabel("Distance (parsec)",size=20)

bar = fig.colorbar(image,orientation="vertical",pad=0.01)
bar.set_label('$log_{10} (Age)$',size=15) #age color bar might be redunant for the age plot, but 
```
